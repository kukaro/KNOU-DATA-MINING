```r
wine <- read.csv('../ch1/data/winequalityCLASS.csv')
```

```r
t(wine)
```

```r
t(wine$quality)
```

```r
wine$quality <- factor(wine$quality)
```

```r
t(wine$quality)
```

```r
library(rpart)
```

```r
set.seed(1234)
```

```r
my.control <- rpart.control(xval=10, cp=0, minsplit=20)
```

```r
my.control
```

```r
tree.wine <- rpart(quality ~ ., data=wine, method='class', control=my.control)
```

```r
tree.wine
```

```r
library(rpart.plot)
```

```r
prp(tree.wine, type=4, extra=1, digits=2, box.palette='Grays')
```

```r
cps <- printcp(tree.wine)
```

```r
t(cps)
```

```r
#which.min은 최소값(0.588342441)의 위치를 의미함.
k <- which.min(cps[,'xerror'])
k
```

```r
err <- cps[k, 'xerror'];
err
```

```r
se <- cps[k, 'xstd']
se
```

```r
c <- 1 # 1 - s.e
```

```r
cps[,'xerror'] <= err + c * se
k1 <- which(cps[,'xerror'] <= err + c * se)[1]
k1
```

```r
cp.chosen <- cps[k1, 'CP']
cp.chosen
```

```r
tree.pruned.wine <- prune(tree.wine, cp=cp.chosen)
```

```r
tree.pruned.wine
```

```r
prp(tree.pruned.wine, type=4, extra=1, digits=2, box.palette='Grays')
```

```r
tree.pruned.wine4 <- prune(tree.wine, cp=cps[4, 'CP'])
prp(tree.pruned.wine4, type=4, extra=1, digits=2, box.palette='Grays')
```
